```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from lcls_beamline_toolbox.xraybeamline2d import beam1d as beam, optics1d as optics, beamline1d as beamline
```

### Beam parameters

```python
# number of sampling points
N = 2048*8
# photon energy in eV
E0 = 1100

# parameter dictionary. z_source is in LCLS coordinates (20 meters upstream of undulator exit)
beam_params = {
    'photonEnergy': E0,
    'N': N,
    'sigma_x': 25e-6,
    'sigma_y': 25e-6,
    'rangeFactor': 5,
    'scaleFactor': 10,
    'z0x': 105.6817,
    'z0y': 106.6817,
    'z_source': 630
}
```

### Define beamline

```python

# initialize the beam
b1 = beam.Beam(beam_params=beam_params)

pf1k0 = optics.WFS('pf1k0', pitch=41e-6, z=735.6817)
im2k0 = optics.PPM('im2k0', FOV=2e-3, blur=True, resolution=5e-6, N=1024, z=pf1k0.z+.76)

# list of devices to propagate through
devices = [pf1k0, im2k0]

# initialize beamline
K_beamline = beamline.Beamline(devices)
```

### Propagation through monochromator at central photon energy

```python
# propagate
b2 = K_beamline.propagate_beamline(b1)

K_beamline.im2k0.view_beam()
# wfs_data = L_beamline.im2l0.retrieve_wavefront(L_beamline.pf1l0)

```
